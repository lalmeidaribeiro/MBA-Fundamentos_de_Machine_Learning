```python
#Tratando a base de dados

del base['Unnamed: 0'] #Deleta coluna
```

```python
base.head()
```

```python
#Separar o que é entrada(x) e o que é saída(y)

#Seleciona todas as linhas da base, mas não todas as colunas
#Função iloc[:, :-1]  [:, ] Seleciona todas as linhas - [,-1] - Informa sobre as colunas exceto a última

x = base.iloc[:,:-1]
```

```python
x
```

```python
#Informando a saída

y = base.iloc[:,-1]
```

```python
y
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Criando a instância, um objeto a partir do classificador

knn = KNeighborsClassifier(n_neighbors=3) #Definindo o hiperparametro. n_neighbors representa o K
```

```python
#Treinando o algoritmo
knn.fit(x,y)  #fit sempre será usada na etapa de treinamento
```

```python
amostra = base.sample() #sample seleciona um elemento de forma aleatória
```

```python
amostra
```

```python
amostra.iloc[:, :-1]
```

```python
knn.predict(amostra.iloc[:, :-1]) #predoct função para prever a linguagem do funcionário
```
