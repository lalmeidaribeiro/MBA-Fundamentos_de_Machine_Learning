```python
import pandas as pd
```

```python
base = pd.read_csv(f'{caminho_arquivo}/compras_supermercado.csv') #sep='' infomro o separador do arquivo
```

```python
base.shape
```

```python
base.head()
```

```python
from sklearn.cluster import KMeans
```

```python
#criando a instância do kmenas
kmeans  = KMeans(n_clusters=4, max_iter=1000)  #n_clusters=4 Número de grupos a serem encontrados max_inter= - Máximo de interações
```

```python
kmeans.fit(base)
```

```python
print(kmeans.get_params()) #Mostra os hiperparâmetros
```

```python
kmeans.labels_ #Qual a atribuição de cada objeto
```

```python
kmeans.cluster_centers_#Identfifcar o valor dos centroides
```

```python
pd.DataFrame(kmeans.cluster_centers_, columns =base.columns) #columns =base.columns pega o nome da base que está sendo analisada
```

```python
import numpy as np
```

```python
grupo, frequencia = np.unique(kmeans.labels_, return_counts = True)
```

```python
print(grupo) #Vejo os grupos
print(frequencia) #Vejo a feequencia dos dados distribuida dentro de cada grupo
```
