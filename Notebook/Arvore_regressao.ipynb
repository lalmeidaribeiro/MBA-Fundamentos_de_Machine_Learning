```python
caminho_arquivo = '/content/drive/MyDrive/Fundamentos_ML_Curso/Bases/'
```

```python
import pandas as pd
```

```python
base = pd.read_excel(f'{caminho_arquivo}/jogadores.xlsx')
```

```python
base.head()
```

```python
del base ['Unnamed: 0']
base
```

```python
x = base.iloc[:,:-1]
y = base.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree = DecisionTreeRegressor(min_samples_leaf=2)
```

```python
tree.fit(x,y)
print(tree.get_params()) #Acessa os hiperparâmetros
```

```python
amostra = base.sample()
```

```python
amostra
```

```python
tree.predict(amostra.iloc[:,:-1])
```

```python
tree.predict([[1,0,1,0,1,1]])
```

```python
from sklearn.tree import export_graphviz
```

```python
arquivo = open("Arvore_regressao.dot","w")
```

```python
export_graphviz(tree, out_file=arquivo, feature_names=x.columns)
```
